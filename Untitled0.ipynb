```python
import conceptnet_lite

conceptnet_lite.connect("./conceptnet.db")
```

```python
from conceptnet_lite import Label, edges_for

cat_concept = Label.get(text='cat').concepts[0]
print(cat_concept)

for edge in edges_for([cat_concept], same_language=True):
    relation = edge.relation.name  # Tên quan hệ (ví dụ: is_a, has_property)
    start_node = edge.start.text   # Từ bắt đầu
    end_node = edge.end.text       # Từ kết thúc

    print(f"{start_node} --({relation})--> {end_node}")
```
